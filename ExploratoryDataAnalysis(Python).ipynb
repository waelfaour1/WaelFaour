```python
# I want to exclude only non-numeric columns
numeric_columns = df.select_dtypes(include=[np.number])

#I have Selected the columns from '1970 Population' to '2022 Population'
selected_columns = df[['1970 Population','1980 Population', '1990 Population','2000 Population', '2010 Population', '2015 Population','2020 Population','2022 Population' ]]

#I have Calculated the mean for selected columns and sort by "2022 Population"
result = df.groupby('Continent')[selected_columns.columns].mean().sort_values(by="2022 Population", ascending=False)
result
```
