```python
df['project_counts']=df['project_name'].map(df['project_name'].value_counts())
df['project_counts']
```

```python
sns.set_palette("husl")


plt.figure(figsize=(12, 6))
sns.barplot(x='project_name', y='project_counts', data=df)

plt.xticks(rotation=90)

plt.xlabel('Project Name')
plt.ylabel('Count')
plt.title('Histogram of Project Counts')


plt.tight_layout()

plt.show()
```
